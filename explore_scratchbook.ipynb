```python
# Users that while active in the curriculum, accessed the curriculum pages less than 100 times
active_records_all.user_id.value_counts()[active_records_all.user_id.value_counts() < 100]
```

```python
# Saving the user_ids that showed low activity while active to a list for iteration
low_access_list = active_records_all.user_id.value_counts()[active_records_all.user_id.value_counts() < 100].index.to_list()
```

```python
cohorts_represented = {}

for user in low_access_list:
    user_data = active_records_all[active_records_all.user_id == user]
    print(f'USER ID: {user}') # Prints user ID
    print(f'COHORT ID: {user_data.cohort_id.value_counts().index[0]}') # Prints cohort id (confirmed that there are no users with more than one cohort id or cohort name)
    print(f'COHORT NAME: {user_data.name.value_counts().index[0]}') # Prints cohort name
    print('\n')
    print('PAGES VISITED:')
    print(user_data.page_viewed.value_counts()) # Prints the page and number of times it was visited
    print('\n')
    print('IP ADDRESSES USED:')
    print(user_data.ip.value_counts()) # Prints the number of times each ip was used
    print('\n')
    print('IP INFORMATION:')
    user_ips = [] # Generates an empty list to store the unique id's for the user
    for i in range(len(user_data.ip.value_counts())): # Iterates a number of times equal to the number of ips the user has
        user_ips.append(user_data.ip.value_counts().index[i]) # Appends the unique ip addresses to the user_ips list
    for ip in user_ips: # Iterates through each up and uses a free geolocation api to retrieve geographic data about the ip
        url = "https://freegeoip.app/json/" + ip
        headers = {
            'accept': "application/json",
            'content-type': "application/json"
            }
        response = requests.request("GET", url, headers=headers)
        print(response.text) # Prints the geolocation data for the ip addressed currently being looped through
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    print('\n')

    if user_data.name.value_counts().index[0] in cohorts_represented:
        cohorts_represented[user_data.name.value_counts().index[0]] += 1 # Adds to the count for each cohort in the cohorts_represented dictionary
    else:
        cohorts_represented[user_data.name.value_counts().index[0]] = 1 # If the dictionary does not yet contain the cohort, creates a new key:value pair

print(cohorts_represented) # Prints the number of times each cohort was represented from the low_access_list subgroup
```

## A Moment to Reconsider

Looking at indivdual users is a time-consuming and potentially low yield process. If there is web scraping involved, we would see a sudden spike from a single user within a short time frame. That single user would then show a lot of webpages being viewed once (maybe twice) as they cycled through the curriculum. We can build a iterative loop that can look for instances like these.

```python
com_full
```

```python
# STEP ONE: Develop a list of user ids
user_ids = list(com_full.user_id.unique())
len(user_ids)
```

```python
# STEP TWO: Iterate through each user id
# We want to look at a resampled dataframe and take note of the one minute window that had the highest level of activity for each user
# We will keep track of these counts in a dictionary
user_max_activity_count = {}
for user in user_ids:
    high_activity_count = com_full[com_full.user_id == user].resample('1min').page_viewed.count().max()
    user_max_activity_count[user] = high_activity_count
```

```python
# We will create a dataframe out of that dictionary
user_max_activity_df = pd.DataFrame.from_dict(user_max_activity_count, orient='index').reset_index().rename(columns={'index': 'user_id', 0:'frequency'})
user_max_activity_df
```

```python
user_max_activity_df.frequency.mean()
```

We will use the interquartile range to identify outliers

```python
def get_lower_and_upper_bounds(s, m = 6):
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower_fence = q1 - m * iqr
    upper_fence = q3 + m * iqr
    return lower_fence, upper_fence
```

```python
get_lower_and_upper_bounds(user_max_activity_df.frequency)
```
